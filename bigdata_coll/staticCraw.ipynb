```python
url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
data = {'stnId': '159'}  
r=requests.get(url,params=data)
r.encoding = 'utf-8'
webPage=r.text
soup=bs(webPage,'html.parser')
print(soup.prettify)
```

```python
locs = soup.find_all("location")
list1 = []

for loc in locs:
    city = loc.city.text
    datas = loc.find_all("data")
    
    for data in datas:
        tmef = data.tmef.text   # TMEF : 예측시각 년월일시 (UTC)
        wf = data.wf.text   # weather forecast
        tmn = data.tmn.text # temperature minimum
        tmx = data.tmx.text # temperature maximum
        list1.append([city, tmef, wf, tmn, tmx])

    print(list1)
```

```python
import pandas as pd
df=pd.DataFrame(list1,columns=['city','tmef','wf','tmn','tmx'])
df
```

title(제목), author(저자), publish(출판사), price(가격), grade(평점)

```python
# 책 목록 선택
book_list = []
books = soup.select('#yesSchList > li')

for book in books:
    # 제목 추출
    title = book.select_one("a.gd_name")
    if title is not None : 
        title_txt=title.text

    # 저자 추출
    author = book.select_one("span.authPub.info_auth > a")
    if author is not None :
        author_txt=author.text

    # 출판사 추출
    publish = book.select_one("span.authPub.info_pub")
    if publish is not None :
        publish_txt=publish.text

    # 가격 추출
    price = book.select_one("strong > em")
    if price is not None :
        price_txt=int(price.text.replace(",",""))

    # 평점 추출
    grade = book.select_one("span.rating_grade > em")
    grade_txt=0.0
    if grade!=None:
        grade_txt=float(grade.text)

    #img=item.select_one("img")
    #img_src = img['data-original']
    #save_image(img_src, title_txt)
    
    book_list.append((title_txt,author_txt,publish_txt,price_txt,grade_txt))

    # 출력
    #print(f"제목: {title_txt}, 저자: {author_txt}, 출판사: {publish_txt}, 가격: {price_txt}, 평점: {grade_txt}")
book_list
```
