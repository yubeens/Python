```python
#기상청 육상 중기 예보 실습을 requests 라이브러리를 사용하여, 경남지역의 일기예보를 출력

import requests

api = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"

# 지역번호 : 전국 108, 서울/경기 109, 강원 105, 충북 131, 충남 133, 전북 146, 전남 156, 경북 143, 경남 159, 제주 184
value = {'stnId': '159'}  
params = requests.compat.urlencode(value)

# 요청 전용 URL 생성
url = api + "?" + params

print("url=", url)

# 다운로드
response = requests.get(url)
response.encoding = 'utf-8'
text = response.text
print(text)
```

```python
# urllib 라이브러리를 사용하여 데이터 get 방식으로 로드하여 출력 

import requests

# URL 설정
url = "http://search.daum.net/nate"
data = {'q': '신지아'}

# GET 요청 보내기
r = requests.get(url, params=data)
r.encoding = 'utf-8'

# 응답 내용 출력
print(r.text)
```
