```python
import urllib.request

res=urllib.request.urlopen('http://www.naver.com')
print(type(res))
print(res.status)
print(res.version)
headers=res.getheaders()
for header in headers:
    print(header)

body=res.read()
print(body.decode('utf-8'))
```

```python
from urllib.request import urlopen
res=urlopen('http://www.naver.com')
print(res.read().decode('utf-8'))
```

```python
import urllib.request as req
res=req.urlopen('http://www.naver.com')
print(res.read().decode('utf-8'))
```

```python
# 예외발생
from urllib.error import URLError, HTTPError
try:
    res=req.urlopen('http://www.naver.com/none')
except HTTPError as e:
    print(e.code)
    print(e.reason)
except URLError as e:
    print(e.reason)
    print('URLError')
else:
    print('Success')
```

```python
url='http://www.naver.com'
res=req.urlopen(url)
a=res.info()
print(type(a))
encoding=a.get_content_charset()
print(encoding)
print(res.read().decode(encoding))
```

```python
print("================================================================================================")
url='http://www.python.org/'
f=urllib.request.urlopen(url)
print(type(f))
print(type(f.info()))
encoding=f.info().get_content_charset()
print(url,'페이지의 인코딩 정보:',encoding)
text=f.read(500).decode(encoding)
print(text)
print("================================================================================================")

url='http://www.google.com/'
f=urllib.request.urlopen(url)
encoding=f.info().get_content_charset()
print(url,'페이지의 인코딩 정보:',encoding)
text=f.read(500).decode(encoding)
print(text)
print("================================================================================================")
```
