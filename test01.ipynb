```python
#관계연산
a=10
b=5

print(a>b)
print(a<b)
print(a>=b)
print(a<=b)
print(a==b)
print(a!=b) 
```

```python
#비트연산
a=8
b=5

print(a&b) #비트곱  1000 & 0101 = 0000
print(a|b) #비트합  1000 | 0101 = 1101
print(a^b) #배타적 논리합 1000 ^ 0101 = 1101
print(~a)  #비트부정  1000 -> 0111
print(a<<2) #왼쪽 시프트 1000 -> 100000
print(a>>2) #오른쪽 시프트 1000 -> 10
print(~a+1) #2의 보수

c=a<<2
print(c) #32 8*4
print(a>>2) #2  8/4
```

```python
#논리연산
print((a>5) and (b<5)) #Flase
print((a>5) or (b<5))  #True
print(not((a>5) and (b<5))) #True
print(not((a>5) or (b<5)))  #False
```
