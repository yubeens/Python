```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
#Matplotlib 기본적인 그래프

#데이터 생성
def chart1(): 
    x=[1,2,3,4,5]
    y=[10,20,25,30,40]

    plt.figure(figsize=(4,4), dpi=128) #그래프 크기와 해상도 설정

    #그래프 그리기
    plt.plot(x,y) #x,y축을 그래프로 그림

    #그래프 제목 및 레이블 추가
    plt.title("Simple Line Graph") #그래프 제목
    plt.xlabel("X Axis") #x축 제목
    plt.ylabel("Y Axis") #y축 제목
    plt.savefig("data/chart1.png") #그래프를 이미지로 저장

    #그래프 출력
    plt.show() 

chart1() 
```

```python
#여러개의 선 그리기

x=[1,2,3,4,5]
y1=[10,20,25,30,40]
y2=[5,15,20,25,30]

plt.plot(x,y1, label='Line 1', color='blue', linestyle='--', marker='o')
plt.plot(x,y2, label='Line 2', color='red', linestyle='-', marker='s')
plt.title("Multiple Line Graphs")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend() #범례 추가
plt.grid(True) #격자 추가
plt.savefig("data/mult_chart.png")
plt.show()
```

```python
#bar차트 그리기

categories=['A','B','C','D']
values=[10,25,15,25]
plt.bar(categories, values, color=['blue','red','green','purple'])
```

```python
#bar차트 그리기2

categories=['A','B','C','D']
values1=[10,25,15,25]
values2=[15,12,20,15]
plt.bar(categories, values1, color=['blue','red','green','purple'])
plt.bar(categories, values2, color=['red','blue','pink','yellow'], bottom=values1) #bottom을 이용해 두번째 bar차트를 첫번째 bar차트 위에 그림
```

```python
#Histogram(히스토그램) 그리기

data=np.random.randn(1000) #정규분포를 따르는 난수 1000개 생성

plt.hist(data, bins=10, color='blue', alpha=0.7) #bins는 막대의 개수, alpha는 불투명도

plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
#Scatter Plot(산포도) 그리기

x=[1,2,3,4,5]
y=[10,15,17,18,21]
colors=np.random.rand(5) #색상을 랜덤으로 생성
sizes=np.random.rand(5)*1000#크기를 랜덤으로 생성
plt.scatter(x,y, c=colors, s=sizes, alpha=0.5, cmap='viridis') #cmap - 색상맵
plt.colorbar() #색상 바 추가
```

```python
#Scatter Plot(산포도) 그리기2

x=np.random.rand(50)
y=np.random.rand(50)
colors=np.random.rand(50)
sizes=np.random.rand(50)*500

plt.scatter(x,y, c=colors, s=sizes, alpha=0.5, cmap='viridis')

plt.title("Scatter Plot Example")
plt.xlabel("X Value")
plt.ylabel("Y Value")

plt.colorbar()
plt.show()
```

```python
#Pie Chart(파이차트) 그리기
labels=['A','B','C','D']
sizes=[10,20,30,40]
colors=['blue','green','red','purple']
explode=(0,0.1,0,0) #두 번째 조각 강조 (분리)

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', explode=explode, startangle=90)
#autopct : 각 섹션에 퍼센트 표시
# startangle: 파이 차트를 90도에서 시작

plt.title("Pie Chart Example")

plt.show()
```

```python
#서브플롯(Subplots) 그리기
import matplotlib.font_manager as fm
plt.rc('font',family='Malgun Gothic')
plt.rc('axes',unicode_minus=False) #유니코드를 적용하면 깨지기 때문에 마이너스는 따로 지정해줘야함 (한글이랑 set로 사용한다고 보면 됨)
#내용에 한글을 기입하려면 따로 폰트매니저를 지정해야함.(안하면 깨짐)

x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)

plt.figure(figsize=(10,5)) #그래프 크기 설정

#첫 번째 그래프 (사인 함수)
plt.subplot(2,1,1) #(행,열,위치)
plt.plot(x,y1, label='sin', color='red')
plt.title("사인함수") 
plt.legend()

#두 번째 그래프 (코사인 함수)
plt.subplot(2,1,2)
plt.plot(x,y2, label='cos', color='blue')
plt.title("코사인 함수")
plt.legend()

plt.tight_layout() #그래프 간격 조정
plt.show()
```

```python
#온도통계 차트 실습!

#자료를 불러와 데이터정보 확인
data=pd.read_csv("data/temperature_2014.csv")
data.info()
```

```python
#PST를 날짜형식으로 변경
data['PST']=pd.to_datetime(data['PST'])
data.info()
```

```python
#한글,마이너스 유니코드 지정
plt.rc('font',family='Malgun Gothic') 
plt.rc('axes',unicode_minus=False)

fig=plt.figure(dpi=128, figsize=(10,6)) #차트 사이즈 지정

#리스트 
pst=list(data['PST'])
high=list(data['Max TemperatureF'])
low=list(data['Min TemperatureF'])

#최저온도 그리기
plt.plot(pst, low, c='blue', alpha=0.5)
#최대온도 그리기
plt.plot(pst, high,c='red', alpha=0.5)
#채우기
plt.fill_between(pst,high, low, facecolor='green',alpha=0.2)

plt.title("온도 통계", fontsize=20)
plt.xlabel("기간",fontsize=16)
plt.ylabel("온도",fontsize=16)
plt.tick_params(axis='both',labelsize=12) #눈금설정 : axis='both' - x축과 y축 모두에 적용
plt.savefig("data/temp_chart.png")
plt.show()
```
