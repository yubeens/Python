```python
import pandas as pd
```

```python
#df 생성

name=['허준호','이가원','배규민','고고림','이새봄','이보람','이루리','오다현']
gender=['남자','여자','남자','남자','여자','여자','여자','여자']
age=[30,24,23,21,28,26,24,24]
hight=[183,162,179,182,160,163,157,172]
data={'이름':name,'성별':gender,'나이':age,'키':hight}
df=pd.DataFrame(data);df
```

```python
#몸무게 추가

df['몸무게']=[82.5,50.0,78.0,80.5,48.0,51.2,46.0,65.0];df
```

```python
#키가 180이상인 사람의 이름과 키를 조회

df.loc[df['키']>=180,['이름','키']]
```

```python
#여자 중 키가 160이상인 사람을 조회

df.loc[(df['성별']=='여자')&(df['키']>160),['이름','키']]
```

```python
#성별 별 키의 평균을 구하기

df.groupby('성별')['키'].mean()
```

```python
#성별 별 키와 몸무게의 평균,합,인원수를 구하라

df.groupby('성별').agg({'키':['mean','sum','count'],'몸무게':['mean','sum','count']})
```

```python
#이름과 성별 칼럼을 추출하여 df2를 생성

df2=df[['이름','성별']];df2
```

```python
#df2에 [연락처] 칼럼을 추가한다

df2['연락처']=['010-1111-2525','010-2222-2525','010-3333-2525','010-4444-2525','010-5555-2525','010-6666-2525','010-7777-2525','010-8888-2525'];df2
```

```python
#df와 df2의 이름 칼럼을 인덱스로 설정

df.set_index('이름',inplace=True)
df2.set_index('이름',inplace=True)
```

```python
df2
```

```python
#df와 df2를 인덱스를 기준으로 조인 

df_join=df.merge(df2,left_index=True, right_index=True);df_join
```

```python
#데이터프레임을 csv파일(test_df.csv)로 저장

df_join.to_csv('data/test_df.csv')
```

```python
#저장된 test_df.csv파일을 데이터프레임으로 로드

df_load=pd.read_csv('data/test_df.csv',index_col=0);df_load
```
