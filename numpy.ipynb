```python
data=np.loadtxt('data/chocolate_rating.csv',delimiter=',') #정수와 실수가 섞여있는 데이터는 실수로 변환
print(data)
print('차원의 수:',data.ndim)
print('모양:',data.shape)
print('원소의 수:',data.size)
```

```python
#초콜릿으로 평점
ratings_mean=data[:,3].mean()
print("모든 초콜릿의 평점:",ratings_mean)

high_level=data[data[:,3]>=4] #평점이 4점 이상인 초콜릿
#평점이 4점 이상인 초콜릿의 개수
print(high_level.shape[0]) 
#print(high_level.shape[1])

high_id=high_level[:,0].astype(np.int64) #평점이 4점 이상인 초콜릿의 id
#print(high_id)
print(f'우수 초콜릿의 수: {len(high_id)}개')
```

```python
#카카오 함유량 빈도수
high_kakao=high_level[:,2] #평점이 4점 이상인 초콜릿의 카카오 함유량
unique_values, value_count=np.unique(high_kakao, return_counts=True) #고유값과 개수를 반환
t=list(zip(unique_values,value_count)) #zip함수를 이용해 고유값과 개수를 묶음
print(t) #고유값과 개수 출력
#print(unique_values, value_count) #고유값과 개수 출력
for k,v in t:
    print(f'카카오 함유량이 {k}인 초콜릿의 수: {v}개')
```

```python
# 카카오 함유량 빈도수가 가장 높은 함유량
aa=np.array([1,5,22,6])
print(np.argmax(aa)) # np.argmax(): 배열에서 가장 큰 값이 있는 위치(인덱스) 리턴
print(value_count)

max_index=np.argmax(value_count) #빈도수가 가장 높은 함유량의 인덱스
print(max_index)
max_kakao=unique_values[max_index] #빈도수가 가장 높은 함유량 출력
print(f'우수 초콜릿 {len(high_id)}가지 중 카카오 함유량이 {max_kakao}인 초콜릿이 가장 많습니다. (빈도 {value_count[max_index]} 개)')
```

```python
data=np.genfromtxt('data/ratings.dat', delimiter='::', dtype=np.int64) #정수 데이터는 정수로 변환
data
```

```python
print(data.shape)
```

```python
print(data[:,0]) #전체 사용자의 id
print(data[:,1]) #전체 영화의 id
print(data[:,2]) #전체 영화의 평점
print(data[:,2].mean()) #전체 영화의 평점
```

```python
#사용자별 영화 평점의 평균
#영화별 영화평점의 평균


user_ids=np.unique(data[:,0]) #고유 사용자 id    
print(len(user_ids)) #고유 사용자 수

mean_values=[] #평점의 평균값을 저장할 리스트
for user_id in user_ids:
    data_user=data[data[:,0]==user_id,:] #사용자별 데이터
    #print(data_user)
    value=data_user[:,2].mean() #사용자별 평점의 평균값
    mean_values.append([user_id,value]) #사용자별 평점의 평균값을 리스트에 추가

mean_values[:5] #테스트 확인용

#data[data[:,0]==1,:] #사용자 id가 1인 데이터
```

```python
print(len(mean_values)) #사용자별 평점의 평균값의 개수
mean_array=np.array(mean_values) #리스트를 배열로 변환
print(mean_array.shape) #배열의 형태
np.savetxt('data/result.csv',mean_array, fmt='%.1f',delimiter=',') #배열을 csv파일로 저장
```
